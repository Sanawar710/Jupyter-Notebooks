# Importing The Desired Libraries
We are importng the libraries that we need to perform calculations, draw graphs and access the data from the file.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
```

# Mounting the Drive
We are mounting the google drive here and accessing the file (on which we are working).

```python
drive.mount('/content/drive') # Mounting the drive
file_path = '/content/drive/MyDrive/DS-221-Assignment/exercise_angles.csv'  # Importing the desired file
data = pd.read_csv(file_path)

data.head() # Displaying the information of first 5 elements
```

# Frequencies

Here, we are calculating frequencies. In our case, we have calculated frequecnies like cumulative frequency and relative frequency.

```python
freq_table = data['Shoulder_Ground_Angle'].value_counts() # Calculates the frequency for 'Shoulder Ground Angle'
freq_table
```

Here, we are plotting the bar plot for the frequency caclulated above.

```python
sns.barplot(x = freq_table.index, y = freq_table.values) # It is used to plot the bar chart
# Setting the titles and labels for x and y axis
plt.title('Frequency Table of Shoulder_Angle')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.show()
```

# Cumulative Frequency
Here, cumulative frequency is calculated. The cumulative frequency is the sum of all the frequencies till a certain point.

```python
cumulative_freq = data['Knee_Angle'].value_counts().cumsum() # Calculates the frequency for 'Knee Angle'
cumulative_freq
```

We are plotting a bar plot for cumulative frequency.

```python
sns.barplot(x=cumulative_freq.index, y=cumulative_freq.values) # It is used to plot the bar chart
# Setting the titles and labels for x and y axis
plt.title('Cumulative Frequency Plot')
plt.xlabel('Knee_Angle')
plt.ylabel('Cumulative Frequency')
plt.show()
```

# Relative Frequency
Here, we are calculating relative frequency. It is the ratio of occurrences relative to the total frequency.

```python
# Calculating Relative Frequency
frequency = data['Knee_Ground_Angle'].value_counts()
relative_freq = frequency / frequency.sum()
relative_freq
```

We are plotting a bar plot for relative frequency.

```python
sns.barplot(x=relative_freq.index, y=relative_freq.values) # It is used to plot the bar chart
# Setting the titles and labels for x and y axis
plt.title('Frequency Table of Knee_Ground_Angle')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.show()
```
