#Importing Essential Libraris
In this section, we have imported the esstenial libraries for a simple linear regression model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

#Generating Data
In this section, we have used the numpy library to generate syntethic data (on a random basis) and added noise.

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y_true = 3 * X
noise = np.random.randn(100, 1)
y = y_true + noise

# Add a bias term (intercept) to X
X_b = np.c_[np.ones((100, 1)), X]
```

# Implementing the Algorithm for Linear Regression
In this section, we have written a simple alogrithm for Single Variable Linear Regression.

```python
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

# The first element of theta_best is the intercept (bias), and the second is the slope
intercept_closed_form = theta_best[0, 0]
slope_closed_form = theta_best[1, 0]

# Make predictions using the closed-form solution
y_predict_closed_form = X_b @ theta_best
```

#Data Visualization
In this section, we have plotted a figure for the linear regression (y-axis vs x-axis).

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Synthetic Data')
plt.plot(X, y_predict_closed_form, color='red', label=f'Closed-Form Regression Line (y = {slope_closed_form:.2f}x + {intercept_closed_form:.2f})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Closed-Form Solution')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the R² score
r2 = r2_score(y, y_predict_closed_form)
print(f"\nR² score using the closed-form solution: {r2:.4f}")
```
